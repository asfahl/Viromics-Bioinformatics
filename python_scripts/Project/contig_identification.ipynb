```python
import pandas as pd
```

```python
# load the quality summary form checkV
checkV = pd.read_csv("C:/Users/David/Desktop/Studium/Viromics-Bioinformatics/Taxonomy/viral_contigs_all_outputs_merged.csv")
checkV = checkV[checkV["prediction"]=="phage"]
```

```python
# sort by ratio between predicted viral gene count and predicted gene count
sort = checkV.sort_values(by="genomad_total_genes", key=lambda col: checkV["genomad_viral_genes"]/col)
```

```python
# only extract high quality reads
sort = sort[sort["checkv_quality"]=="High-quality"].reset_index()
sort.head(20)
```

```python
diff = checkV.sort_values(by="genomad_total_genes", key=lambda col: col - checkV["genomad_viral_genes"], ascending=False).reset_index()
```

```python
diff = diff[diff["checkv_quality"]=="High-quality"]
diff.head(12)
```
